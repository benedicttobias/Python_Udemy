```python
from geopy.geocoders import Nominatim
```

```python
# Build object
nom = Nominatim(scheme="http")
```

```python
# Get the data
nom.geocode("250 Brent Lane, Pensacola, FL 32503")
```

```python
# Get the data (invalid address)
print(nom.geocode("123 Semoga Awet raya"))
```

```python
# Store to variable and use it
pccAddress = nom.geocode("250 Brent Lane, Pensacola, FL 32503")
pccAddress.latitude, pccAddress.longitude
```

```python
import pandas
df = pandas.read_csv("supermarkets.csv")
df
```

```python
df["Address"] = df["Address"] + ", " + df["City"] + ", " + df["State"] + ", " + df["Country"]
df
```

```python
# Iterate without for loop -- Applying geocode
# Build object
df["Coordinates"] = df["Address"].apply(nom.geocode)
df
```

```python
df.Coordinates
```

```python
# Add latitude to the column
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude)

# Causing error due to one "NONE" on the row
```

```python
# Fixed: Add latitude to the column
# Store none if no latitude
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"] = df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df
```
